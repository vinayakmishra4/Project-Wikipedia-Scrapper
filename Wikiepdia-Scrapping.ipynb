# Wikiepdia Scrapping

```python
# import libarray
import requests as req
from bs4 import BeautifulSoup
```

```python
title=str(input("Enter the title of the Wikipedia page you want to scrape: "))
url=f"https://en.wikipedia.org/wiki/{title.replace(' ', '_')}"
print(f"Scraping data from: {url}")
```

```python
# make the request to the website and get the HTML content
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 ""(KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"}
response=req.get(url, headers=headers)
print(response.status_code)
```

```python
# parse the HTML content and extract the desired information
soup = BeautifulSoup(response.content, 'html.parser')
headings = soup.find('h1').text
print(headings)
```

```python
# extract all the paragraphs from the webpage and print them
for p in soup.find_all('p'):
    print(p.text)
    print("-"*10)
```

```python
# save the extracted data to a file
file_name = f"{title.replace(' ', '_')}.txt"
with open(file_name, 'w', encoding='utf-8') as file:
    file.write(headings + "\n\n")
    for p in soup.find_all('p'):
        file.write(p.text + "\n\n")
```
